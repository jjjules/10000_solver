```python
import pandas as pd

data = [
    ['filtres Jules', 4, 16.4, 'jules'], ['feuilles', 4, 29.8, 'jules'], ['feuilles', 2, 29.8, 'aenis'], ['feuilles', 1, 29.8, 'theo'],
    ['poppers', 1, 6.5, 'theo'], ['poppers', 1, 14.4, 'jules'], ['filtres JASS', 10, 4.5, 'hugo'], ['filtres JASS', 10, 4.5, 'aenis'], ['filtres JASS', 1, 4.5, 'theo'],
    ['longue feuilles JASS', 2, 22.9, 'hugo'], ['filtres normaux', 1, 49.9, 'jules'],
    ['longue feuilles JASS', 2, 22.9, 'jules'],
]

entries = pd.DataFrame(data=data, columns=['item_name', 'num_items', 'item_price', 'person'])
entries['total_price'] = entries.num_items * entries.item_price
display(entries)

per_person = entries.groupby('person').total_price.sum()
per_person
```
